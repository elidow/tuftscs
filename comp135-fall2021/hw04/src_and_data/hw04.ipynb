#### Import required libraries

```python
import os
import numpy as np
import pandas as pd
import time
import warnings

from sklearn.neural_network import MLPClassifier

from matplotlib import pyplot as plt
import seaborn as sns

from MLPClassifierWithSolverLBFGS import MLPClassifierLBFGS

from viz_tools_for_binary_classifier import plot_pretty_probabilities_for_clf

%matplotlib inline
```

#### Load data

```python
# Load data
x_tr_N2 = np.loadtxt('data_xor/x_train.csv', skiprows=1, delimiter=',')
x_te_N2 = np.loadtxt('data_xor/x_test.csv', skiprows=1, delimiter=',')

y_tr_N = np.loadtxt('data_xor/y_train.csv', skiprows=1, delimiter=',')
y_te_N = np.loadtxt('data_xor/y_test.csv', skiprows=1, delimiter=',')

assert x_tr_N2.shape[0] == y_tr_N.shape[0]
assert x_te_N2.shape[0] == y_te_N.shape[0]
```

### Problem 1: MLP size [2] with activation ReLU and L-BFGS solver

```python
# setting number of runs/models to 16
num_runs = 16 

def train_MLP_with_RELU_and_LBFGS(total_runs):
    ''' Trains MLPClassifier models (# = total_runs), each one using a different random intitization-state.
        Models will have two hidden nodes that use the ReLU activation, and will be optimized with the L-BFGS solver
    
    Args
    ----
    total_runs : integer value
        Represents the number of times we train a model with a different random intitization-state

    Returns
    -------
    MLP_classifiers: list of MLP classifiers
        Each entry represents a classifier for a trained MLP model
        Each entry has a different random intitization-state
    '''
    
    # initialize local variables
    MLP_classifiers = []
    n_runs = total_runs
    
    # training models (# = n_runs)
    for i in range(n_runs):
        
        # training model with random_state i and timing it
        start_time_sec = time.time()
        mlp_lbfgs = MLPClassifierLBFGS(
            hidden_layer_sizes=[2],
            activation='relu',
            alpha=0.0001,
            max_iter=200, tol=1e-6,
            random_state=i,
        )
        with warnings.catch_warnings(record=True) as warn_list:
            mlp_lbfgs.fit(x_tr_N2, y_tr_N)
        elapsed_time_sec = time.time() - start_time_sec
        
        # printing results of run
        print('finished LBFGS run with RELU activation  %2d/%d after %6.3f sec | %3d iters | %s | loss %.3f' % (
            (i+1), n_runs, elapsed_time_sec,
            len(mlp_lbfgs.loss_curve_),
            'converged   ' if mlp_lbfgs.did_converge else 'NOT converged',
            mlp_lbfgs.loss_))
        
        # adding to trained MLP classifiers
        MLP_classifiers.append(mlp_lbfgs)
    
    return MLP_classifiers
    

# training MLP models with RELU activation and L-BFGS solver
mlp_relu_lbfgs_models = train_MLP_with_RELU_and_LBFGS(num_runs)
```

#### 1 (a): Visualize probabilistic predictions in 2D feature space for ReLU + L-BFGS

```python
# initializing 2D feature space for ReLU + L-BFGS as a 4 by 4 grid 
fig, ax_grid = plt.subplots(nrows=4, ncols=4, figsize=(16, 16))

# populating grid with each model's probabilistic predictions
for i in range(num_runs):
    plot_pretty_probabilities_for_clf(mlp_relu_lbfgs_models[i], x_tr_N2, y_tr_N, ax=ax_grid[i // 4, i % 4])
```

### Problem 3: MLP size [2] with activation ReLU and SGD solver

```python
def train_MLP_with_RELU_and_SGD(total_runs):
    ''' Trains MLPClassifier models (# = total_runs), each one using a different random intitization-state.
        Models will have two hidden nodes that use the ReLU activation, and will be optimized with the SGD solver
    
    Args
    ----
    total_runs : integer value
        Represents the number of times we train a model with a different random intitization-state

    Returns
    -------
    MLP_classifiers: list of MLP classifiers
        Each entry represents a classifier for a trained MLP model
        Each entry has a different random intitization-state
    '''
    
    # initialize local variables
    MLP_classifiers = []
    n_runs = total_runs

        
    # training models (# = n_runs)
    for i in range(n_runs):
        
        # training model with random_state i and timing it
        start_time_sec = time.time()
        mlp_sgd = MLPClassifier(
            hidden_layer_sizes=[2],
            activation='relu',
            alpha=0.0001,
            solver='sgd', batch_size=10,
            max_iter=400, tol=1e-8,
            learning_rate='adaptive', learning_rate_init=0.1, momentum=0.0,
            random_state=i,
            )
        with warnings.catch_warnings(record=True) as warn_list:
            mlp_sgd.fit(x_tr_N2, y_tr_N)
        elapsed_time_sec = time.time() - start_time_sec
        
        # setting did_converge attribute appropriately to avoid errors (true if there are no warnings)
        if len(warn_list) == 0:
            mlp_sgd.did_converge = True
        else:
            mlp_sgd.did_converge = False
        
        # printing results of run
        print('finished SGD run with RELU activation  %2d/%d after %6.1f sec | %3d iters | %s | loss %.3f' % (
            (i+1), n_runs, elapsed_time_sec,
            len(mlp_sgd.loss_curve_),
            'converged    ' if mlp_sgd.did_converge else 'NOT converged',
            mlp_sgd.loss_))

        # adding to trained MLP classifiers
        MLP_classifiers.append(mlp_sgd)
    
    return MLP_classifiers
    

# training MLP models with RELU activation and SGD solver
mlp_relu_sgd_models = train_MLP_with_RELU_and_SGD(num_runs)
```

#### 3 (a): Visualize probabilistic predictions in 2D feature space for ReLU + SGD

```python
# initializing 2D feature space for ReLU + SGD as a 4 by 4 grid 
fig, ax_grid = plt.subplots(nrows=4, ncols=4, figsize=(16, 16))

# populating grid with each model's probabilistic predictions
for i in range(num_runs):
    plot_pretty_probabilities_for_clf(mlp_relu_sgd_models[i], x_tr_N2, y_tr_N, ax=ax_grid[i // 4, i % 4])
```
